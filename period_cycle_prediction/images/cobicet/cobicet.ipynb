```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils.utils import generate_synthetic_data, generate_final_features, split_dataset
import warnings 
warnings.filterwarnings("ignore")
```

```python
# load the data
cycle_data = pd.read_csv('dataset\\synthetic_data.csv')
```

```python
# generate the final features
features_cycle_data, labels_regular_cycle_data = generate_final_features(cycle_data)
```

```python
# slice the data with split_dataset function
input_train_cycle, input_test_cycle, output_train_cycle, output_test_cycle = split_dataset(features_cycle_data, labels_regular_cycle_data)
```

```python
#plot the distribution of the data with distplot
sns.distplot(tuple(input_train_cycle[:,0]), hist=True, kde=True, kde_kws={'color': 'red', 'shade': False, 'linewidth': 3}, label='Total Regular Cycle')

# change font size of the labels 
plt.rcParams.update({'font.size': 20})
plt.xlabel('Durao dos ciclos (dias)')
plt.ylabel('Densidade')
fig = plt.gcf()
fig.savefig('cobicet_fig1.png', dpi=300, bbox_inches='tight')
```

```python
#plot the distribution of the data with distplot
sns.distplot(tuple(input_train_cycle[:,1]), hist=True, kde=True, kde_kws={'color': 'red', 'shade': False, 'linewidth': 3}, label='Total Regular Cycle')

# change font size of the labels 
plt.rcParams.update({'font.size': 20})
plt.xlabel('Durao do perodo (dias)')
plt.ylabel('Densidade')
fig = plt.gcf()
fig.savefig('cobicet_fig2.png', dpi=300, bbox_inches='tight')
```
